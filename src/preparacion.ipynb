```python
import pandas as pd

```

```python
writer = pd.ExcelWriter('data/vinos.xlsx', engine='xlsxwriter')
```

```python
pro_var = pd.read_excel(r'vinos.xlsx', sheet_name='produccion_variedades')


pro_var
```

```python
columnas = pro_var.columns
```

```python

columnas = list(map(str.lower,columnas))
```

```python
columnas[0] = 'año'
```

```python
pro_var.columns = columnas
```

```python
pro_var.head()
```

```python
pro_var = pro_var.melt(id_vars=["año"], 
        var_name="variedad", 
        value_name="produccion")
```

```python
pro_var.head()
```

```python
pro_var.to_excel(writer, sheet_name='pro_var', index = False)
```
